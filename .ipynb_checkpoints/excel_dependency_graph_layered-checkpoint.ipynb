
# Visualizing Excel Model Logic with Python

In the previous post we built a small **Excel auditing script** that scanned a workbook for structural issues like:

• hardcoded constants  
• inconsistent formulas  
• overwritten values  

In this notebook we take a different perspective.

Instead of looking for errors, we visualize **how the spreadsheet actually works**.

Excel models are really **networks of dependencies**. Every formula references other cells, and those relationships form a directed graph.

Using Python and NetworkX we will:

• extract formula references  
• build a dependency graph  
• expand Excel ranges like `E2:E4`  
• visualize the workbook structure  
• arrange the graph so it mirrors the model layout

## Import libraries

```python

import openpyxl
import networkx as nx
import matplotlib.pyplot as plt
import re
from openpyxl.utils.cell import coordinate_from_string, column_index_from_string
from openpyxl.utils import get_column_letter
```

## Load the workbook

```python

file = "demo_financial_model.xlsx"

wb = openpyxl.load_workbook(file, data_only=False)

print("Sheets:", wb.sheetnames)
```


## Helper function: expand Excel ranges

Excel formulas often reference ranges such as:

`E2:E4`

We expand these into individual cells:

`E2, E3, E4`


```python

def expand_range(sheet, start, end):

    col1, row1 = coordinate_from_string(start)
    col2, row2 = coordinate_from_string(end)

    col1_i = column_index_from_string(col1)
    col2_i = column_index_from_string(col2)

    cells = []

    for col in range(col1_i, col2_i + 1):
        for row in range(row1, row2 + 1):

            coord = f"{get_column_letter(col)}{row}"
            cells.append(f"{sheet}!{coord}")

    return cells
```

## Build the dependency graph

```python

G = nx.DiGraph()

cell_pattern = r"[A-Za-z_]+!\$?[A-Za-z]+\$?\d+"
range_pattern = r"([A-Za-z_]+)!([A-Za-z]+\d+):([A-Za-z]+\d+)"

for sheet in wb.sheetnames:

    ws = wb[sheet]

    for row in ws.iter_rows():

        for cell in row:

            if cell.data_type == "f":

                target = f"{sheet}!{cell.coordinate}"
                formula = cell.value

                # Expand ranges
                ranges = re.findall(range_pattern, formula)

                for r in ranges:
                    rng_sheet, start, end = r
                    expanded = expand_range(rng_sheet, start, end)

                    for ref in expanded:
                        G.add_edge(ref, target)

                # Capture individual references
                refs = re.findall(cell_pattern, formula)

                for ref in refs:
                    ref = ref.replace("$","")
                    G.add_edge(ref, target)

print("Nodes:", len(G.nodes))
print("Edges:", len(G.edges))
```

## Color nodes by worksheet

```python

def node_color(node):

    if node.startswith("Inputs"):
        return "lightgreen"

    elif node.startswith("Calculations"):
        return "lightskyblue"

    elif node.startswith("Outputs"):
        return "salmon"

    else:
        return "lightgray"


colors = [node_color(n) for n in G.nodes]
```


## Create a layered layout

Instead of using a physics layout, we position nodes by worksheet:

Inputs → left  
Calculations → middle  
Outputs → right

This mirrors how analytical models are typically structured.


```python

layers = {
    "Inputs": 0,
    "Calculations": 1,
    "Outputs": 2
}

pos = {}
y_positions = {}

for node in G.nodes:

    sheet = node.split("!")[0]
    layer = layers.get(sheet, 3)

    y_positions.setdefault(sheet, 0)

    pos[node] = (layer, -y_positions[sheet])
    y_positions[sheet] += 1
```

## Visualize the dependency graph

```python

plt.figure(figsize=(10,8))

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color=colors,
    node_size=2000,
    font_size=8,
    arrows=True
)

plt.title("Excel Model Dependency Graph (Layered Layout)")
plt.show()
```


### Color legend

Green → Inputs  
Blue → Calculations  
Red → Outputs  

Now the graph mirrors the structure of the Excel model.


## Impact analysis example

```python

input_cell = "Inputs!B4"

affected = nx.descendants(G, input_cell)

print("Cells affected by", input_cell)

for c in affected:
    print(c)
```

## Circular reference detection

```python

cycles = list(nx.simple_cycles(G))

if cycles:

    print("Circular references detected:")

    for c in cycles:
        print(c)

else:
    print("No circular references detected")
```
