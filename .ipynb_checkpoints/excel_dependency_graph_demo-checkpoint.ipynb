
# Visualizing Excel Model Logic with Python

In the previous post we built a simple **Excel auditing script** that scans workbooks for structural issues like hardcoded constants and formula inconsistencies.

In this walkthrough we take a different perspective.

Instead of looking for potential errors, we visualize **how the spreadsheet actually works**. Excel models are really networks of dependencies where formulas reference other cells. That means we can represent a workbook as a **directed graph**.

Using Python and NetworkX we can:

- Extract formula relationships
- Build a dependency graph
- Visualize how calculations flow through the workbook
- Identify important or fragile cells in the model


## 1. Install required libraries

## 2. Import libraries

```python

import openpyxl
import networkx as nx
import matplotlib.pyplot as plt
import re
```

## 3. Load the Excel workbook

```python

file = "demo_financial_model.xlsx"

wb = openpyxl.load_workbook(file, data_only=False)

print("Sheets in workbook:", wb.sheetnames)
```

## 4. Build the dependency graph

```python

G = nx.DiGraph()

cell_pattern = r"[A-Za-z_]+!\$?[A-Za-z]+\$?\d+|\$?[A-Za-z]+\$?\d+"

for sheet in wb.sheetnames:
    
    ws = wb[sheet]
    
    for row in ws.iter_rows():
        
        for cell in row:
            
            if cell.data_type == "f":
                
                target = f"{sheet}!{cell.coordinate}"
                formula = cell.value
                
                refs = re.findall(cell_pattern, formula)
                
                for ref in refs:
                    
                    if "!" not in ref:
                        ref = f"{sheet}!{ref}"
                    
                    ref = ref.replace("$","")
                    
                    G.add_edge(ref, target)
```

## 5. Graph statistics

```python

print("Number of cells in graph:", len(G.nodes))
print("Number of dependencies:", len(G.edges))
```

## 6. Color nodes by worksheet

```python

def node_color(node):

    if node.startswith("Inputs"):
        return "lightgreen"
    
    elif node.startswith("Calculations"):
        return "lightskyblue"
    
    elif node.startswith("Outputs"):
        return "salmon"
    
    else:
        return "lightgray"


colors = [node_color(n) for n in G.nodes]
```

## 7. Visualize the dependency graph

```python

plt.figure(figsize=(12,8))

pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color=colors,
    node_size=2000,
    font_size=8,
    arrows=True
)

plt.title("Excel Model Dependency Graph")
plt.show()
```


### Color legend

- Green → Inputs sheet  
- Blue → Calculation sheet  
- Red → Output sheet  

This makes it easy to see the architecture of the model.


## 8. Impact analysis

```python

input_cell = "Inputs!B4"

affected = nx.descendants(G, input_cell)

print("Cells affected by", input_cell)

for c in affected:
    print(c)
```

## 9. Detect circular references

```python

cycles = list(nx.simple_cycles(G))

if cycles:
    print("Circular references detected:")
    
    for c in cycles:
        print(c)
        
else:
    print("No circular references detected")
```
