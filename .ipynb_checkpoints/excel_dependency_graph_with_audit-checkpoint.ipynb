
# Visualizing Excel Model Logic with Python (with Audit Highlights)

In the previous post we built a simple **Excel auditing script** that detected:

- hardcoded constants
- overwritten values
- inconsistent formulas

In this notebook we extend that idea.

Instead of only printing audit results, we **overlay those findings on a dependency graph** of the spreadsheet.  
This allows us to see **where problems appear inside the model structure**.

We will:

- extract Excel formula dependencies
- expand cell ranges like `E2:E4`
- build a dependency graph
- layer the graph Inputs → Calculations → Outputs
- highlight hardcoded calculation cells

## Import libraries

```python

import openpyxl
import networkx as nx
import matplotlib.pyplot as plt
import re
from openpyxl.utils.cell import coordinate_from_string, column_index_from_string
from openpyxl.utils import get_column_letter
```

## Load workbook

```python

file = "demo_financial_model.xlsx"

wb = openpyxl.load_workbook(file, data_only=False)

print("Sheets:", wb.sheetnames)
```

## Detect hardcoded constants in calculation sheets

```python

hardcoded_cells = set()

input_sheets = ["Inputs"]

for sheet in wb.sheetnames:
    
    ws = wb[sheet]
    
    for row in ws.iter_rows():
        
        for cell in row:
            
            if cell.data_type == "n":
                
                if sheet not in input_sheets:
                    
                    if cell.value is not None:
                        
                        hardcoded_cells.add(f"{sheet}!{cell.coordinate}")
                        
print("Hardcoded calculation cells:", hardcoded_cells)
```

## Helper function to expand Excel ranges

```python

def expand_range(sheet, start, end):

    col1, row1 = coordinate_from_string(start)
    col2, row2 = coordinate_from_string(end)

    col1_i = column_index_from_string(col1)
    col2_i = column_index_from_string(col2)

    cells = []

    for col in range(col1_i, col2_i + 1):
        for row in range(row1, row2 + 1):

            coord = f"{get_column_letter(col)}{row}"
            cells.append(f"{sheet}!{coord}")

    return cells
```

## Build the dependency graph

```python

G = nx.DiGraph()

cell_pattern = r"[A-Za-z_]+!\$?[A-Za-z]+\$?\d+"
range_pattern = r"([A-Za-z_]+)!([A-Za-z]+\d+):([A-Za-z]+\d+)"

for sheet in wb.sheetnames:

    ws = wb[sheet]

    for row in ws.iter_rows():

        for cell in row:

            if cell.data_type == "f":

                target = f"{sheet}!{cell.coordinate}"
                formula = cell.value

                ranges = re.findall(range_pattern, formula)

                for r in ranges:
                    
                    rng_sheet, start, end = r
                    expanded = expand_range(rng_sheet, start, end)

                    for ref in expanded:
                        G.add_edge(ref, target)

                refs = re.findall(cell_pattern, formula)

                for ref in refs:
                    
                    ref = ref.replace("$","")
                    G.add_edge(ref, target)

print("Nodes:", len(G.nodes))
print("Edges:", len(G.edges))
```

## Color nodes based on role and audit findings

```python

def node_color(node):

    if node in hardcoded_cells:
        return "gold"

    if node.startswith("Inputs"):
        return "lightgreen"

    elif node.startswith("Calculations"):
        return "lightskyblue"

    elif node.startswith("Outputs"):
        return "salmon"

    else:
        return "lightgray"

colors = [node_color(n) for n in G.nodes]
```

## Create layered layout (Inputs → Calculations → Outputs)

```python

layers = {
    "Inputs": 0,
    "Calculations": 1,
    "Outputs": 2
}

pos = {}
y_positions = {}

for node in G.nodes:

    sheet = node.split("!")[0]
    layer = layers.get(sheet, 3)

    y_positions.setdefault(sheet, 0)

    pos[node] = (layer, -y_positions[sheet])

    y_positions[sheet] += 1
```

## Visualize the dependency graph with highlighted issues

```python

plt.figure(figsize=(10,8))

sizes = [500 + len(nx.descendants(G,n))*150 for n in G.nodes]

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color=colors,
    node_size=sizes,
    font_size=8,
    arrows=True
)

plt.title("Excel Model Dependency Graph with Audit Highlights")
plt.show()
```


### Color legend

Green → Inputs  
Blue → Calculations  
Red → Outputs  
Gold → Hardcoded constants in calculation sheets

This allows us to visually identify structural violations in the model.


## Impact analysis example

```python

input_cell = "Inputs!B4"

affected = nx.descendants(G, input_cell)

print("Cells affected by", input_cell)

for c in affected:
    print(c)
```

## Detect circular references

```python

cycles = list(nx.simple_cycles(G))

if cycles:

    print("Circular references detected:")

    for c in cycles:
        print(c)

else:
    print("No circular references detected")
```
